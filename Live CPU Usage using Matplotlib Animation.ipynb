```python
import random
from itertools import count
import pandas as pd
from psutil import *
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation 
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
x_vals = [0,1,2,3,4,5]
y_vals = [0,1,3,2,3,5]
plt.plot(x_vals, y_vals)
plt.tight_layout()
plt.show()
plt.close()
```

```python
%matplotlib notebook
x_vals = []
y_vals = []
index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(cpu_percent(1, percpu=True)[0])
    plt.cla()
    plt.plot(x_vals, y_vals)

ani = FuncAnimation(plt.gcf(), animate, interval=1000)
plt.tight_layout()
plt.show()
```

```python
%matplotlib notebook
x_vals = []
y_vals = []
index = count()
plt.figure(figsize=(10, 5))
def animate(i):
    data = pd.read_csv('data.csv')
    x = data['x']
    y1 = data['CPU-1']
    y2 = data['CPU-2']
    y3 = data['CPU-3']
    y4 = data['CPU-4']
    
    plt.cla()
    plt.plot(x, y1, label='CPU-1')
    plt.plot(x, y2, label='CPU-2')
    plt.plot(x, y3, label='CPU-3')
    plt.plot(x, y4, label='CPU-4')
    
    plt.legend(loc = 'upper left')

ani = FuncAnimation(plt.gcf(), animate, interval=1000)
plt.tight_layout()
plt.show()
```
